
## Some Visualizations done on the Dataset 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import re
%matplotlib inline
```

```python
data= pd.read_csv("GrievancesData_Train.csv")
data
```

```python
##Plotting the column variable

plt.figure()
data.State.value_counts().plot(kind='bar',
                                  figsize=(20,10),
                                  color="blue",
                                  alpha = 0.7,
                                  fontsize=13)
plt.title('Checking the Total Number of States')
plt.grid()
plt.show()
```

```python
##Observations:- 

#We see that State 9 has the highest level of Grievance management in the order of things followed by State 14.
```

```python
data.State.value_counts()
```

```python
##Checking the number of the Grievance_Category of the Dataset

plt.figure()
data.Grievance_Category.value_counts().plot(kind='bar',
                                  figsize=(20,10),
                                  color="blue",
                                  alpha = 0.7,
                                  fontsize=13)
plt.title('Grievance_Category of the Dataset')
plt.grid()
plt.show()
```

```python
##Observations:-

# From the above plot we can infer that the category of "Loan Servicing,payments,escrow account" has the highest distribution
# in the dataset followed by "Loan modification,Collection,Foreclosure" which suggests that the majority of the time is involved
# in these activities.
```

```python
###Generating WordCloud for the Dataset

from wordcloud import WordCloud
wordcloud = WordCloud().generate(str(data["GrievanceDescription"]))
import matplotlib.pyplot as plt
%matplotlib inline
plt.imshow(wordcloud,interpolation='bilinear')
# interpolation="bilinear" in the plt.imshow() is to make the displayed image appear more smoothly


plt.figure( figsize=(15,10) )
plt.imshow(wordcloud)
```

```python
## Observations:- 

# WordCloud gives us the result of the most frequent words occuring the the Dataset.
```

```python
## Plotting another Plot for Line of Business

plt.figure()
data.LineOfBusiness.value_counts().plot(kind='bar',
                                  figsize=(20,10),
                                  color="blue",
                                  alpha = 0.7,
                                  fontsize=13)
plt.title('Line Of Business of the Dataset')
plt.grid()
plt.show()
```

```python
##Observations:-

# We see that CreditCard has the highest Line Of business followed by Mortgage
```

```python
## Plotting another graph for the ResolutionComments category

plt.figure()
data.ResolutionComments.value_counts().plot(kind='bar',
                                  figsize=(20,10),
                                  color="blue",
                                  alpha = 0.7,
                                  fontsize=13)
plt.title('Checking the Distribution of ResolutionComments')
plt.grid()
plt.show()
```

```python
## Observations:-

# Plot results indicate that most of the cases has been closed with Explanation followed by Monetary Relief.
```

```python
## Ploting the Distribution of Disputed Closure in the Dataset

plt.figure()
data.Disputed.value_counts().plot(kind='bar',
                                  figsize=(20,10),
                                  color="blue",
                                  alpha = 0.7,
                                  fontsize=13)
plt.title('Checking the Distribution of Disputed Closure in the Dataset')
plt.grid()
plt.show()
```

```python
## Observations:-

# We see most of the Disputed Data in the Dataset is NO which implies that Customer's is not happy with the resolution provided 
# by the Bank.
```
